```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Cargar los datos desde el archivo
data = np.loadtxt("resultados.dat")

# Separar en variables x e y
x_vals = data[:, 0]
y_vals = data[:, 1]

# Graficar los datos
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label="Solución RK4", color="b")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Método de Runge-Kutta de 4to Orden")
plt.legend()
plt.grid()
plt.show()
```
